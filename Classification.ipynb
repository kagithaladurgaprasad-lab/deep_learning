```python
#drop 
df.drop(["sample_id","harvest_date"],axis=1,inplace=True)      
```

```python
df.shape
```

```python
#EDA 
#preprocessing 
```

```python
 #target variable quality category
y=df['quality_category']
```

```python
x =df.drop(['quality_category'],axis=1)
```

```python
x.shape
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,StandardScaler
le=LabelEncoder()
y=le.fit_transform(y)
print(le.classes_)
```

```python
x.drop(["firmness_kgf","variety","weight_g","altitude_m","soil_nitrogen_ppm"],axis=1,inplace=True)
```

```python
#apply columntransformer 
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer(transformers=[("oe",OrdinalEncoder(),[0,3,]),
                                   ("sc",StandardScaler(),[1,2,4,5,6,7])])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train_transform=ct.fit_transform(x_train)
x_train_transform=pd.DataFrame(x_train_transform)
```

```python
x_train_transform.shape
```

```python
x_test_transform=ct.fit_transform(x_test)
x_test_transform=pd.DataFrame(x_test_transform)
```

# import libraries

```python
from tensorflow.keras.models import Sequential
```

```python
from keras.layers import Dense,BatchNormalization
```

# iniatlize

```python
model=Sequential()
```

## add layers

```python
# input layer + hidden layes
model.add(Dense(units=64,activation="relu",input_shape=(x_train_transform.shape[1],)))
model.add(BatchNormalization())
## hiden layer 2
model.add(Dense(units=10,activation="relu"))
model.add(BatchNormalization())
## output layer
model.add(Dense(units=4,activation="softmax"))
```

```python
model.summary()
```

# compile model

```python
model.compile(optimizer="sgd",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

 # Train Model

```python
hist=model.fit(x_train_transform,y_train,epochs=50,verbose=1,validation_split=0.2,batch_size=32)
```

# evalute

```python
acc=model.evaluate(x_test_transform,y_test)
```

```python
import matplotlib.pyplot as plt

#Extract metrics from the history object
acc = hist.history['r2_score']
val_acc = hist.history['val_r2_score']
loss = hist.history['loss']
val_loss = hist.history['val_loss']
epochs = range(1, len(acc) + 1)

#Plot Accuracy
plt.figure(figsize=(10, 4))
plt.title("Results with adam")
plt.subplot(1, 2, 1)
plt.plot(epochs, acc, 'bo-', label='Training Accuracy')
plt.plot(epochs, val_acc, 'ro-', label='Validation Accuracy')
plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

#Plot Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, loss, 'bo-', label='Training Loss')
plt.plot(epochs, val_loss, 'ro-', label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
