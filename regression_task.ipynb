```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.datasets import load_iris
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score ,classification_report 
```

```python
df=pd.read_csv("Advertising.csv")
```

```python
df.isnull().sum()
```

```python
df.drop("Unnamed: 0",axis=1)
```

```python
x=df.drop("Sales",axis=1)
y=df['Sales']
```

```python
# train test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
x_train_tran=mn.fit_transform(x_train)
```

```python
x_train_tran=pd.DataFrame(x_train_tran)
```

```python
x_test_tran=mn.transform(x_test)
x_test_tran=pd.DataFrame(x_test_tran)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
from tensorflow.keras.layers import Dense, BatchNormalization
```

```python


#add input layer 
model.add(Dense(64,activation='relu',input_shape=(x_train_tran.shape[1],)))
model.add(BatchNormalization())

#hidden layer 
model.add(Dense(32,activation="relu"))
model.add(BatchNormalization())

#add output layer 
model.add(Dense(1))
```

```python
import tensorflow as tf
```

```python
#compile the model
model.compile(optimizer='adam',loss='mse',metrics=[tf.keras.metrics.R2Score()])
```

```python
price=model.fit(x_train_tran,y_train,epochs=30,batch_size=1,verbose=1)
```

```python
r2=model.evaluate(x_test_tran,y_test)
```

```python
import matplotlib.pyplot as plt

#Extract metrics from the history object
acc = price.history['r2_score']
val_acc = price.history['val_r2_score']
loss = price.history['loss']
val_loss = price.history['val_loss']
epochs = range(1, len(acc) + 1)

#Plot Accuracy
plt.figure(figsize=(10, 4))
plt.title("Results with adam")
plt.subplot(1, 2, 1)
plt.plot(epochs, acc, 'bo-', label='Training Accuracy')
plt.plot(epochs, val_acc, 'ro-', label='Validation Accuracy')
plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

#Plot Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, loss, 'bo-', label='Training Loss')
plt.plot(epochs, val_loss, 'ro-', label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
